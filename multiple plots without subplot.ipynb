```python
import pandas as pd, matplotlib.pyplot as plt, seaborn as sns
```

```python
df  = sns.load_dataset('tips')
df
```

```python
fig = plt.figure(figsize =(12,12))
# violinplot
a = fig.add_axes([0.1, 0.6, 0.3, 0.3])
sns.violinplot(df['total_bill'],ax = a, orient = 'h')

#histplot
b = fig.add_axes([0.5, 0.6, 0.3, 0.3])
sns.histplot(df['total_bill'],ax = b, kde = True)

# mean, median
c = fig.add_axes([0.9, 0.6, 0.3, 0.3])
sns.kdeplot(df['total_bill'],  ax = c)
median = df['total_bill'].median()
mean = df['total_bill'].mean()
std_dev = df['total_bill'].std()
for i in range(1, 4):
    plt.axvline(median + i*std_dev, color='gray')
    plt.axvline(median - i*std_dev, color='blue')
    plt.axvline(mean+i*std_dev, color = 'black')
    plt.axvline(mean - i*std_dev,color = 'green')


#scatterplot
d = fig.add_axes([0.1, 0.3, 0.3, 0.25])
sns.scatterplot(data = df, x = 'total_bill', y = 'tip', hue = 'smoker', ax = d)

#lineplot
e = fig.add_axes([0.5, 0.3, 0.3, 0.25])
sns.lineplot(data = df, x = 'total_bill',y = 'tip', ax = e)


#swarnplot
g = fig.add_axes([0.9, 0.3, 0.3, 0.25])
sns.swarmplot(data = df, x = 'total_bill', y = 'tip',hue = 'day')



```
